```
df = pd.read_excel("./drive/My Drive/data/food.xlsx")
```

```
df.restaurent=df.restaurent.replace("eatfit","eat fit")
```

```
df['restaurent']=df['restaurent'].str.title()
```

```
df.restaurent.count()
```

```
df.restaurent.nunique()
```

```
df.restaurent.value_counts()
```

```
dic=df.restaurent.value_counts()
```

```
dic
```

```
wordcloud = WordCloud(
    width = 2000,
    height = 1000,repeat=False,
    background_color = 'black')
```

```
wordcloud.generate_from_frequencies(dic)
                                    
```

```
fig = plt.figure(
    figsize = (10, 5),
    facecolor = 'k',
    edgecolor = 'k')
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

```
wordcloud.to_file("./drive/My Drive/data/food.png")
```
